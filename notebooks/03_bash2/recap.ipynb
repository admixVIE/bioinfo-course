## Working with commands
Looking at things:

`ls`, `less`, `cat`, `wc`

What does each of these commands do?

Moving & deleting:

`cp`, `mv`, `rm`

**Loops:**

`for` bla `in` range
`do` command 
`done`

**Subsetting:**

`grep "pattern" file` - getting lines of a file which include specific patterns

`-v` - do the opposite!

`cut -f1-2 -d " " file` - getting specific columns of a file (counting from first column)

`awk 'pattern {subset}'` - more complex subsetting using a small programming language

`sort` for sorting and `uniq` for unique rows

**Regular expressions:**

Allows more complex patterns to be found. Within context of *pattern* parameters of programs such as `grep`.

Some are the same symbols/characters used elsewhere. You specify the use of regular expressions with e.g. `grep -E`.


**Important things:**

Piping: `|` hands over results from one command to another. Note: within regular expression, the same symbol means `or`!

Commands have *standard input* stream coming in, *standard output* going out, and *standard error* for messages.

Files can be compressed, usually with the file ending `.gz`, `.tar`, or `.tar.gz`. Then, not all tools work the same way anymore.
